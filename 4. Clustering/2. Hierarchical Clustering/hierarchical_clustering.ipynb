# Hierarchical Clustering

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
df = pd.read_csv('/content/Mall_Customers.csv')
X = df.iloc[:, [3, 4]].values
```

## Using the dendrogram to find the optimal number of clusters

```
import scipy.cluster.hierarchy as sch
#'ward' method is for minimized variance
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrograms')
plt.xlabel('Customer Data Points')
plt.ylabel('Euclidean Distance')
plt.show()
```

From here, we can see that 5 clusters is the optimal number of clusters. We have largest distance without cutting any horizontal line with five crossings is there at around (Threshold = 150).

## Training the Hierarchical Clustering model on the dataset

```
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```
print(y_hc)
```

## Visualising the clusters

```
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], color = 'magenta', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], color = 'cyan', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], color = 'green', label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], color = 'red', label = 'Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], color = 'blue', label = 'Cluster 5')
plt.title('Visulaizing Clusters')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending Score(1-100)')
plt.show()
```
